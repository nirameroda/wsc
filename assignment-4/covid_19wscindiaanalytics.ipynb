```
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
#read the dataset

csvdata = pd.read_csv("/content/drive/My Drive/covid19confirmedindia.csv")
print(csvdata)
#print(type(csvdata))
#print(csvdata['Confirmed'])
```

```
covid19= csvdata[csvdata['Date']=="26/04/20"]
print("total confirmed cases")
covid19['Confirmed'].sum()
```

```
confirm = covid19.sort_values(by=['Confirmed'], ascending = False)
plt.figure(figsize=(12,8), dpi=80)
plt.bar(confirm['State'][:5], confirm['Confirmed'][:5],
        align='center',color='green')
plt.ylabel('Number of Confirmed Cases', size = 12)
plt.xlabel('states', size = 12)
plt.title('States with maximum confirmed cases', size = 16)
plt.show()
```

```

deaths = covid19.sort_values(by=['Deaths'], ascending = False)
plt.figure(figsize=(12,8), dpi=80)
plt.bar(deaths['State'][:10], deaths['Deaths'][:10],
        align='center',color='green')
plt.ylabel('Number of deaths', size = 12)
plt.xlabel('states', size = 12)
plt.title('States with maximum deaths cases', size = 16)
plt.show()
print("total deaths")
covid19['Deaths'].sum()
```

```
cured = covid19.sort_values(by=['Cured'], ascending = False)
plt.figure(figsize=(12,8), dpi=80)
plt.bar(cured['State'][:10], cured['Cured'][:10],
        align='center',color='green')
plt.ylabel('Number of patients cured', size = 12)
plt.xlabel('states', size = 12)
plt.title('States with maximum cured cases', size = 16)
plt.show()
print("total Cured")
covid19['Cured'].sum()
```
