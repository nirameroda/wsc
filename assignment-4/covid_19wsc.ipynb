```
def plot_degree_distribution(G):
	all_degree=dict(nx.degree(G)).values()

	unique_degree=list(set(all_degree))
	count_degree=[]
	all_degree=list(all_degree)
	for i in unique_degree:
		x=all_degree.count(i)
		count_degree.append(x)
	
	
	#print(count_degree)
	plt.plot(unique_degree,count_degree,'yo-')
	plt.xlabel('Degrees')
	plt.ylabel('No of nodes ')
	plt.title('Degree distribution of network')
	plt.show()
plot_degree_distribution(G)
```

```
#albert barabasi model
GB=nx.barabasi_albert_graph(27890, 20, seed=None)

degrees=nx.degree(GB)
patients=[]
states=[]
state_degree=dict(degrees)
for i in range(32):
  if(state_degree[i]>1):
    states.append(i)
    patients.append(state_degree[i])
plt.figure(num=None, figsize=(20, 8), dpi=100, facecolor='r', edgecolor='g')
plt.xlabel("States of India")
plt.ylabel("number of patients with covid-19 +ve")
plt.title("Degree distribution of covid19-india confirmed cases till 26-04-2020 ")
plt.bar(states, patients,width = 0.2, color = ['blue'])
#plt.plot(states,patients)
plt.show()
# clustering Cofficient
print("average clustering coefficent")
print(nx.average_clustering(GB))
```

```
#degree centrality of the given graph
dc1=nx.degree_centrality(G)
print("index of max degree centrality",sorted(dc1.keys(), key=lambda key:dc1[key], reverse=True)[:1])
print("index of min degree centrality",sorted(dc1.keys(), key=lambda key:dc1[key], reverse=False)[:1])
plt.plot(list(dc1.keys()),list(dc1.values()),'bo')
plt.xlabel('node index')
plt.ylabel('degree centrality')
```

```
#betweenness centrality of the given graph
d=nx.betweenness_centrality(G)
print(d)

pagemax = max(d, key=d.get)
print('node - '+str(pagemax)+' has maximum betweenness centrality of : '+str(d[pagemax]))

plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.title("Betweenness centrality")
plt.ylabel("nodes")
plt.xlabel("Betweenness centrality")
plt.hist(list(d.values()))
plt.show()
```

```
#closeness centrality
cc1=nx.closeness_centrality(G)
print("index of max closeness centrality",sorted(cc1.keys(), key=lambda key:cc1[key], reverse=True)[:1])
print("index of min closeness centrality",sorted(cc1.keys(), key=lambda key:cc1[key], reverse=False)[:1])
plt.plot(list(cc1.keys()),list(cc1.values()),'bo')
plt.xlabel('node index')
plt.ylabel('closeness centrality')
```

```
#eigenvector centrality
ec1=nx.eigenvector_centrality(G, max_iter=1000, tol=1e-06, nstart=None, weight=None)
print("index of max eigenvector centrality",sorted(ec1.keys(), key=lambda key:ec1[key], reverse=True)[:1])
print("index of min eigenvector centrality",sorted(ec1.keys(), key=lambda key:ec1[key], reverse=False)[:1])
plt.plot(list(ec1.keys()),list(ec1.values()),'bo')
plt.xlabel('node index')
plt.ylabel('eigenvector centrality')
```
