#1.Implement degree centrality,closeness centrality,betweenness centrality and eigen centrality for 3 different graphs discussed in assignment1

1.(a)GRAPH1

```
G1 = nx.read_edgelist("/content/drive/My Drive/p2p-Gnutella09.txt",create_using=nx.DiGraph(), nodetype = int)
print(nx.info(G1))
```

```
dc1=nx.degree_centrality(G1)
print("index of max degree centrality",sorted(dc1.keys(), key=lambda key:dc1[key], reverse=True)[:1])
print("index of min degree centrality",sorted(dc1.keys(), key=lambda key:dc1[key], reverse=False)[:1])
plt.plot(list(dc1.keys()),list(dc1.values()),'bo')
plt.xlabel('node index')
plt.ylabel('degree centrality')
```

```
cc1=nx.closeness_centrality(G1)
print("index of max closeness centrality",sorted(cc1.keys(), key=lambda key:cc1[key], reverse=True)[:1])
print("index of min closeness centrality",sorted(cc1.keys(), key=lambda key:cc1[key], reverse=False)[:1])
plt.plot(list(cc1.keys()),list(cc1.values()),'bo')
plt.xlabel('node index')
plt.ylabel('closeness centrality')
```

```
bc1=nx.betweenness_centrality(G1,normalized=True, weight=None, endpoints=False, seed=None)
print("index of max betweenness centrality",sorted(bc1.keys(), key=lambda key:bc1[key], reverse=True)[:1])
print("index of min betweenness centrality",sorted(bc1.keys(), key=lambda key:bc1[key], reverse=False)[:1])
plt.plot(list(bc1.keys()),list(bc1.values()),'bo')
plt.xlabel('node index')
plt.ylabel('betweenness centrality')
```

```
ec1=nx.eigenvector_centrality(G1, max_iter=100, tol=1e-06, nstart=None, weight=None)
print("index of max eigenvector centrality",sorted(ec1.keys(), key=lambda key:ec1[key], reverse=True)[:1])
print("index of min eigenvector centrality",sorted(ec1.keys(), key=lambda key:ec1[key], reverse=False)[:1])
plt.plot(list(ec1.keys()),list(ec1.values()),'bo')
plt.xlabel('node index')
plt.ylabel('eigenvector centrality')
```

#2.(a)Webpages  have the highest and lowest PageRank and Hubs-Authorities score

#HITS

```
hubs1, authorities1 = nx.hits(G1, max_iter = 50, normalized = True) 
print("Hub Scores: G1 ", hubs1) 
print("Authority Scores: G1 ", authorities1) 
#index of maximum hub score webpage
print("index of maximum hub score webpage",sorted(hubs1.keys(), key=lambda key:hubs1[key], reverse=True)[:1])
#index of minimum hub score webpage
print("index of minimum hub score webpage",sorted(hubs1.keys(), key=lambda key:hubs1[key], reverse=False)[:1])
#index of maximum authority score webpage
print("index of maximum authority score webpage",sorted(authorities1.keys(), key=lambda key:authorities1[key], reverse=True)[:1])
#index of minimum authority score webpage
print("index of minimum authority score webpage",sorted(authorities1.keys(), key=lambda key:authorities1[key], reverse=False)[:1])
```

```
hubs2, authorities2 = nx.hits(G2, max_iter = 50, normalized = True) 
print("Hub Scores:G2 ", hubs2) 
print("Authority Scores:G2 ", authorities2) 
#index of maximum hub score webpage
print("index of maximum hub score webpage",sorted(hubs2.keys(), key=lambda key:hubs2[key], reverse=True)[:1])
#index of minimum hub score webpage
print("index of minimum hub score webpage",sorted(hubs2.keys(), key=lambda key:hubs2[key], reverse=False)[:1])
#index of maximum authority score webpage
print("index of maximum authority score webpage",sorted(authorities2.keys(), key=lambda key:authorities2[key], reverse=True)[:1])
#index of minimum authority score webpage
print("index of minimum authority score webpage",sorted(authorities2.keys(), key=lambda key:authorities2[key], reverse=False)[:1])
```

```
hubs3, authorities3 = nx.hits(G3, max_iter = 100, normalized = True) 
print("Hub Scores:G3 ", hubs3) 
print("Authority Scores:G3 ", authorities3) 
#index of maximum hub score webpage
print("index of maximum hub score webpage",sorted(hubs3.keys(), key=lambda key:hubs3[key], reverse=True)[:1])
#index of minimum hub score webpage
print("index of minimum hub score webpage",sorted(hubs3.keys(), key=lambda key:hubs3[key], reverse=False)[:1])
#index of maximum authority score webpage
print("index of maximum authority score webpage",sorted(authorities3.keys(), key=lambda key:authorities3[key], reverse=True)[:1])
#index of minimum authority score webpage
print("index of minimum authority score webpage",sorted(authorities3.keys(), key=lambda key:authorities3[key], reverse=False)[:1])
```

#page rank

```
pr1 = nx.pagerank(G1, alpha=0.85)
#page rank of all the webpages
print("page rank:G1",pr1)
#index of maximum page rank webpage
print("index of maximum page rank webpage",sorted(pr1.keys(), key=lambda key:pr1[key], reverse=True)[:1])
#index of minimum page rank webpage
print("index of minimum page rank webpage",sorted(pr1.keys(), key=lambda key:pr1[key], reverse=False)[:1])
```
